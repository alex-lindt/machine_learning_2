### Notation

$\mathbf{X}$ is the $M \times T$ data matrix, containing $M$ measurements at $T$ time steps.

$\mathbf{S}$ is the $S \times T$ source matrix, containing $S$ source signal values at $T$ time steps. We will assume $S = M$.

$\mathbf{A}$ is the mixing matrix. We have $\mathbf{X} = \mathbf{A S}$.

$\mathbf{W}$ is the matrix we aim to learn. It is the inverse of $\mathbf{A}$, up to indeterminacies (scaling and permutation of sources).

$\phi$ is an elementwise non-linearity or activation function, typically applied to elements of $\mathbf{W X}$.

````python
%pylab inline
import sys
assert sys.version_info[:3] >= (3, 6, 0), "Make sure you have Python 3.6 installed"

# Signal generators
def sawtooth(x, period=0.2, amp=1.0, phase=0.):
    return (((x / period - phase - 0.5) % 1) - 0.5) * 2 * amp

def sine_wave(x, period=0.2, amp=1.0, phase=0.):
    return np.sin((x / period - phase) * 2 * np.pi) * amp

def square_wave(x, period=0.2, amp=1.0, phase=0.):
    return ((np.floor(2 * x / period - 2 * phase - 1) % 2 == 0).astype(float) - 0.5) * 2 * amp

def triangle_wave(x, period=0.2, amp=1.0, phase=0.):
    return (sawtooth(x, period, 1., phase) * square_wave(x, period, 1., phase) + 0.5) * 2 * amp

def random_nonsingular_matrix(d=2):
    """
    Generates a random nonsingular (invertible) matrix of shape d*d
    """
    epsilon = 0.1
    A = np.random.rand(d, d)
    while abs(np.linalg.det(A)) < epsilon:
        A = np.random.rand(d, d)
    return A

def plot_signals(X, title="Signals"):
    """
    Plot the signals contained in the rows of X.
    """
    figure()
    
    for i in range(X.shape[0]):
        ax = plt.subplot(X.shape[0], 1, i + 1)
        plot(X[i, :])
        ax.set_xticks([])
        ax.set_yticks([])
    plt.suptitle(title)
````

#### Answer:
As apparent from the above plots, the measurement distributions (i.e. the mixtures of the underlying source distributions) tend to look more like Gaussians than the source distributions. This observation is consistent with the central limit theorem, which states that the sum of independent random variables (here: the source variables) tends to have a normal distribution, even if the original variables themselves are not normally distributed.

With ICA, we aim to seperate independent underlying variables of observed (and usually normally distributed) data. This process can be regarded the reverse of the central limit theorem. For the ICA to work, it is necessary to assume that the underlying source variables are not normally distributed. Otherwise, a measured Gaussian distribution could be explained by multiple different mixtures of underlying Gaussian distributions and therefore it would not be possible to obtain the mixture matrix $G$ (or $W = G^{-1}$) or the original sources from it.



### 1.3 Implicit priors (20 points)
As explained in MacKay's book, an activation function $\phi$ used in the ICA learning algorithm corresponds to a prior distribution over sources. Specifically, $\phi(a) = \frac{d}{da} \ln p(a)$. For each of the following activation functions, *derive* the source distribution they correspond to.
$$\phi_0(a) = -\tanh(a)$$
$$\phi_1(a) = -a + \tanh(a)$$
$$\phi_2(a) = -a^3$$
$$\phi_3(a) = -\frac{6a}{a^2 + 5}$$

Give your answer without the normalizing constant, so an answer of the form $p(a) \propto \verb+[answer]+$ is ok.

Generally, we find that 
\begin{align*}
\phi(a) &= \frac{\partial}{\partial a} \ln p(a) \ \ \ \ \ \implies \ \ \ \ \ 
p(a)= \exp\bigg(\int_a \phi(a) da\bigg)\ \ \  \text{.}
\end{align*}
Therefore, we can calculate

(i)
\begin{align*}
p_0(a) = \exp\bigg(\int_a -tanh(a) da\bigg) = \exp\bigg(-ln(cosh(a)+C)\bigg)\propto \frac{1}{cosh(a)}=\frac{2}{(e^a+e^{-a})}=sech(a)
\end{align*}

(ii)
\begin{align*}
p_1(a) &= \exp\bigg(\int_a -a + \tanh(a) da\bigg) =
\exp\bigg(\ln (\cosh (a))-\frac{a^{2}}{2}+C \bigg)
\propto
 \cosh (a)\exp\bigg(-\frac{a^{2}}{2}\bigg)
\end{align*}

(iii)
\begin{align*}
p_2(a) &= \exp\bigg(\int_a -a^3  da\bigg)=
\exp\bigg( \frac{-a^4}{4}+C\bigg)\propto \exp\bigg( \frac{-a^4}{4}\bigg)
\end{align*}

(iv)
\begin{align*}
p_3(a) &= \exp\bigg(\int_a -\frac{6a}{a^2 + 5} da\bigg)=
\exp\bigg(-3 \ln (a^{2}+5) + C \bigg)
\propto
\exp\bigg(\ln (a^{2}+5)\bigg)^{-3} = \frac{1}{(a^{2}+5)^3}
\end{align*}

### 1.4 Whitening (15 points)
Some ICA algorithms can only learn from whitened data. Write a method `whiten(X)` that takes a $M \times T$ data matrix $\mathbf{X}$ (where $M$ is the dimensionality and $T$ the number of examples) and returns a whitened matrix. If you forgot what whitening is or how to compute it, various good sources are available online, such as http://courses.media.mit.edu/2010fall/mas622j/whiten.pdf. Your function should also center the data before whitening.

````python
from numpy import linalg as LA

### 1.4 Whitening
def whiten(X):
    # following steps from http://courses.media.mit.edu/2010fall/mas622j/whiten.pdf.
    # center data
    X = X - np.mean(X, axis=1, keepdims=True)
    # form covariance matrix
#     Sigma = np.cov(X)
    Sigma = (X @ X.T)/X.shape[1]
    
    # solve eigenvalue problem
    _lambda, Phi = LA.eig(Sigma)
    
    Lambda_1 = LA.inv(np.diag(np.sqrt(_lambda)))
    
    # apply whitening transformation
    return (Phi @ Lambda_1).T @ X 
````

### 1.6 Covariance (5 points)
Explain what a covariant algorithm is. 

The information of the follwing are largely based on MacKay's paper [Maximum Likelihood and Covariant Algorithms for Independent Component Analysis](http://www.inference.org.uk/mackay/ica.pdf) and chapter 34.3 of his book (referenced in the beginning of this notebook).

The idea of a covariant algorithm stems from the argument that "a consistent algorithm should give the same results independent of the units in which quantities are measured". A prominent example of a non-covariant algorithm is gradient descent, where dimensional inconsistencies in the update rule arise: 
$$
    \Delta w_i^{(k+1)}  =  w_i^{(k+1)} - \Delta w_i^{(k)}
$$
$$
    \text{where: }\Delta w_i^{(k)} = \eta \frac{\partial L}{\partial w_i^{(k)}},
$$

where we have $w_i^{(k)}$ on the left side of the second equation and $1/w_i^{(k)}$ on the right side. Although gradient descent proves that non-covariant algorithms can work, covariant algorithms promise to converge faster. To transform a non-convariant algorithm into a convariant one, we need to address the dimensional inconsistencies. This can be done by including a positive-definite matrix $M$, where $dim(M_{ii'}) = dim(w_i w_{i'})$. $M$ either originates from the natural distance metric in parameter space of $w$ (if such a natural distance metric exists) or from considering the curvature of the objective function (with $M = (-\nabla \nabla L)^{-1}$, where $\nabla \equiv \partial / \partial w$). A covatiant example of the gradient descent update would be 
$$
    \Delta w_i^{(k)} = \eta \sum_{i'} M_{ii'}\frac{\partial L}{\partial w_i^{(k)}}.
$$

Implement the covariant ICA algorithm as described in MacKay. Write a function `ICA(X, activation_function, learning_rate)`, that returns the demixing matrix $\mathbf{W}$. The input `activation_function` should accept a function such as `lambda a: -tanh(a)`. Update the gradient in batch mode, averaging the gradients over the whole dataset for each update. Make it efficient, so use matrix operations instead of loops where possible (loops are slow in interpreted languages such as python and matlab, whereas matrix operations are internally computed using fast C code). Experiment with the learning rate and the initialization of $\mathbf{W}$. Your algorithm should be able to converge (i.e. `np.linalg.norm(grad) < 1e-5`) in within 10000 steps.

````python
### 1.7 Independent Component Analysis
def ICA(x, activation_function, learning_rate=0.05, verbose=True, return_steps=False):
    
    # obviously, it did not work with lr=1.0 as set in the testing cell below
    learning_rate=0.05
    
    D, N = x.shape
    
    # initialize weights W 
    W =  np.eye(D)
    
    gradient=1e20
    epoch = 0 
    
    while np.linalg.norm(gradient)>1e-5 and epoch<10000:
        
        # following pseudocode in book
        a = W @ x
        z = activation_function(a)
        x_new = W.T @ a
        # update W
        gradient = W + z @ x_new.T /N
        W += learning_rate*gradient
        
        epoch+=1
    
    if verbose:    
        print('ICA DONE.     Epoch %s    Gradient %s\n\n' % (epoch, np.linalg.norm(gradient)))  
    
    if return_steps:
        return epoch
    
    return W
````
