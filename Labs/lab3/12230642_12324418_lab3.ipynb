## Part 1: Expectation Maximization
We will use the Expectation Maximization (EM) algorithm for the recognition of handwritten digits in the MNIST dataset. The images are modelled as a Bernoulli mixture model (see Bishop $\S9.3.3$):
$$
p(\bx|\bmu, \bpi) = \sum_{k=1}^K  \pi_k \prod_{i=1}^D \mu_{ki}^{x_i}(1-\mu_{ki})^{(1-x_i)}
$$
where $x_i$ is the value of pixel $i$ in an image, $\mu_{ki}$ represents the probability that pixel $i$ in class $k$ is black, and $\{\pi_1, \ldots, \pi_K\}$ are the mixing coefficients of classes in the data. We want to use this data set to classify new images of handwritten numbers.

### 1.2 Implementation (40 points)
You are going to write a function ```EM(X, K, max_iter)``` that implements the EM algorithm on the Bernoulli mixture model. 

The only parameters the function has are:
* ```X``` :: (NxD) array of input training images
* ```K``` :: size of the latent space
* ```max_iter``` :: maximum number of iterations, i.e. one E-step and one M-step

You are free to specify your return statement.

Make sure you use a sensible way of terminating the iteration process early to prevent unnecessarily running through all epochs. Vectorize computations using ```numpy``` as  much as possible.

You should implement the `E_step(X, mu, pi)` and `M_step(X, gamma)` separately in the functions defined below. These you can then use in your function `EM(X, K, max_iter)`.

````python
def E_step(X, mu, pi):
    # YOUR CODE HERE
    # In order to vectorize the calculations we have to alter X
    X_aug = X[:, np.newaxis, :]
    
    p_x_mu = np.prod(np.power(mu, X_aug) * np.power(1 - mu, 1 - X_aug), axis=2)
    numerator = pi * p_x_mu
    denominator = np.sum(numerator, axis=1)

    return numerator / denominator[:, np.newaxis]
````

````python
def M_step(X, gamma):
    # YOUR CODE HERE
    N_k = sum(gamma, axis=0)
    mu = (1/N_k[:, np.newaxis]) * np.asarray([np.sum(gamma[:, k, np.newaxis] *  X, axis=0) for k in np.arange(gamma.shape[1])])
    pi = N_k / np.sum(N_k)
    
    return mu, pi
````

````python
def EM(X, K, max_iter, conv_thresh=1e-4, mu=None, pi=None):
    # YOUR CODE HERE
    
    D = X.shape[1]
    
    # Initualize params
    if pi is None:
        pi = np.ones(K)*(1/K)
    if mu is None:
        mu = np.random.uniform(0.25, 0.75, (K, D))
        # Normalization suggested in bishop gives problems
#         mu = mu / np.sum(mu, axis=1)[:, np.newaxis]
        
    log_likelihoods = []
    
    for iteration in np.arange(max_iter):                
        gamma = E_step(X, mu, pi)
        mu, pi = M_step(X, gamma)
        
        llh = np.sum(np.log(
            np.sum(pi * np.prod(np.power(mu, X[:, np.newaxis, :]) * 
                                np.power(1 - mu, 1 - X[:, np.newaxis, :]), axis=2), axis=1)))
        
        log_likelihoods.append(llh)
        
        # log likelihood produces nans...
        if iteration > 0:
            delta_mu = np.linalg.norm(mu - mu_)
            delta_pi = np.linalg.norm(pi - pi_)
            
            print("Iteration: {}, delta_mu: {}, delta_pi: {}, likelihood: {}".format(iteration, np.round(delta_mu, 4), np.round(delta_pi, 4), np.round(llh, 4)))

            if delta_mu < conv_thresh or delta_pi < conv_thresh:
                break
                
        pi_ = pi
        mu_ = mu
    
    return mu, pi, log_likelihoods, gamma
````

````python
def order_predictions(distances, labels, K):
    # Here we want to match the predictions with the acending order of the data labels.
    
    latent_labels = []
    min_distances = np.argmin(distances, axis=0)
    
    for k in np.arange(K):
        mask_idx = np.where(labels == k)
        # here we find the most probable latent label for the cluster
        most_likely_class = np.argmin(distances[:, mask_idx].squeeze(1).sum(axis=1))
        latent_labels.append(most_likely_class)
    
    return np.asarray(latent_labels)[min_distances], latent_labels

def split_data(data, K):
    split_data = []
    for d_sub in np.split(data, K):
        split_data.append(d_sub)
    return split_data

_mu, _, _, _ = EM(training_set, K, 1000)

# Calculate l2 norm between the training data and each mu_k
distance_data_mu = np.asarray([((m - training_set)**2).sum(axis=1) for m in _mu]).reshape(3, 600)
ordered_predictions, latent_labels = order_predictions(distance_data_mu, matching_labels, K)

ordered_predictions_divided = split_data(ordered_predictions, K)
split_labels = split_data(matching_labels, K)
split_train_set = split_data(training_set, K)

wrong_pred_total = np.argwhere(np.not_equal(ordered_predictions, matching_labels))

print("Results: ")
print(" ---- "*10)
print("Total missclassifications: {}".format(wrong_pred_total.shape[0]))
print("Total misclassification rate: {}".format(np.round(wrong_pred_total.shape[0] / n_train, 4)))
print(" ---- "*10)

for k in np.arange(K):
    print(" ---- "*10)
    perdictions = ordered_predictions_divided[k]
    labels = split_labels[k]
    train_imgs = split_train_set[k] 
    wrong_preds = np.argwhere(np.not_equal(perdictions, labels))
    print("Total missclassifications for digit {}: {}".format(k+2, wrong_preds.shape[0]))
    print("Misclassification rate of digit {}: {}".format(k+2, wrong_preds.shape[0] / labels.shape[0]))

    misclassifications_to = perdictions[wrong_preds].squeeze(1)
    misclass_counts = np.bincount(misclassifications_to)
    misclass_counts_idx = np.nonzero(misclass_counts)[0]
    print("In the case of a wrong prediction:")
    for idx in misclass_counts_idx:
        print("Digit {} is confused with digit {} in {} percent of the times".format(k+2, idx+2, np.round(misclass_counts[idx]/wrong_preds.shape[0], 4)))
    
    print(" ---- "*10)
    misclassified_images = train_imgs[wrong_preds].squeeze(1)
    no_images = min(6, misclassified_images.shape[0])
    sample_mci = misclassified_images[:no_images]
    sample_mistakes = misclassifications_to[:no_images]
    
    fig = plt.figure(figsize=(15, 10), dpi=80)
    fig.suptitle("Some hard to classify {}'s".format(k+2), fontweight='bold', fontsize=18, y=0.65)

    for i in np.arange(no_images):
        ax = fig.add_subplot(1, no_images, i+1)
        ax.set_title("Confused with {}".format(sample_mistakes[i]+2))
        ax.imshow(sample_mci[i].reshape((28, 28)))
        
    
````

## Part 2: Variational Auto-Encoder

A Variational Auto-Encoder (VAE) is a probabilistic model $p(\bx, \bz)$ over observed variables $\bx$ and latent variables and/or parameters $\bz$. Here we distinguish the decoder part, $p(\bx | \bz) p(\bz)$ and an encoder part $p(\bz | \bx)$ that are both specified with a neural network. A lower bound on the log marginal likelihood $\log p(\bx)$ can be obtained by approximately inferring the latent variables z from the observed data x using an encoder distribution $q(\bz| \bx)$ that is also specified as a neural network. This lower bound is then optimized to fit the model to the data. 

The model was introduced by Diederik Kingma (during his PhD at the UVA) and Max Welling in 2013, https://arxiv.org/abs/1312.6114. 

Since it is such an important model there are plenty of well written tutorials that should help you with the assignment. E.g: https://jaan.io/what-is-variational-autoencoder-vae-tutorial/.

In the following, we will make heavily use of the torch module, https://pytorch.org/docs/stable/index.html. Most of the time replacing `np.` with `torch.` will do the trick, e.g. `np.sum` becomes `torch.sum` and `np.log` becomes `torch.log`. In addition, we will use `torch.FloatTensor()` as an equivalent to `np.array()`. In order to train our VAE efficiently we will make use of batching. The number of data points in a batch will become the first dimension of our data tensor, e.g. A batch of 128 MNIST images has the dimensions [128, 1, 28, 28]. To check check the dimensions of a tensor you can call `.size()`.

### 2.1 Loss function
The objective function (variational lower bound), that we will use to train the VAE, consists of two terms: a log Bernoulli loss (reconstruction loss) and a KullbackLeibler divergence. We implement the two terms separately and combine them in the end.
As seen in Part 1: Expectation Maximization, we can use a multivariate Bernoulli distribution to model the likelihood $p(\bx | \bz)$ of black and white images. Formally, the variational lower bound is maximized but in PyTorch we are always minimizing therefore we need to calculate the negative log Bernoulli loss and KullbackLeibler divergence.

### 2.1.1 Negative Log Bernoulli loss (5 points)
The negative log Bernoulli loss is defined as,

\begin{align}
loss = - (\sum_i^D \bx_i \log \hat{\bx_i} + (1  \bx_i) \log(1  \hat{\bx_i})).
\end{align}

Write a function `log_bernoulli_loss` that takes a D dimensional vector `x`, its reconstruction `x_hat` and returns the negative log Bernoulli loss. Make sure that your function works for batches of arbitrary size.

````python
def log_bernoulli_loss(x_hat, x):
    # YOUR CODE HERE
    return -torch.sum(x * torch.log(x_hat) + (1 - x) * torch.log(1 - x_hat))
````

### 2.1.2 Negative KullbackLeibler divergence (10 Points)
The variational lower bound (the objective to be maximized) contains a KL term $D_{KL}(q(\bz)||p(\bz))$ that can often be calculated analytically. In the VAE we assume $q = N(\bz, \mu, \sigma^2I)$ and $p = N(\bz, 0, I)$. Solve analytically!

We apply the same procedure as in homework 2 exercise 3.1:
$$
\begin{aligned}
D_{KL}(q(\boldsymbol{z})||p(\boldsymbol{z})) &= \int_{\boldsymbol{z}} q(\boldsymbol{z}) \log \frac{q(\boldsymbol{z})}{p(\boldsymbol{z})} d\boldsymbol{z} \\
    &= \int_{\boldsymbol{z}} q(\boldsymbol{z}) \log \frac{N(\boldsymbol{z}, \mu, \sigma^2I)}{N(\boldsymbol{z}, 0, I)} d\boldsymbol{z}\\
    &= \int_{\boldsymbol{z}} q(\boldsymbol{z}) \bigg( \log N(\boldsymbol{z}, \mu, \sigma^2I) - \log N(\boldsymbol{z}, 0, I) \bigg) d\boldsymbol{z} \\
    &= \int q(\boldsymbol{z}) \bigg( \ln \big( (2\pi)^{-\frac{D}{2}}|(\sigma^2I)|^{-\frac{1}{2}}\exp{\{-\frac{1}{2}(\boldsymbol{z}-\mu)^{T}(\sigma^2I)^{-1}(\boldsymbol{z}-\mu)}\}  \big) \\ &- \ln \big( (2\pi)^{-\frac{D}{2}}\exp{\{-\frac{1}{2}\boldsymbol{z}^{T}\boldsymbol{z} \}} \big) \bigg) d\boldsymbol{z} \\
    &= \int q(\boldsymbol{z}) \bigg( -\frac{D}{2}\ln(2\pi)-\frac{1}{2}\ln(|(\sigma^2I)|)-\frac{1}{2}(\boldsymbol{z}-\mu)^{T}(\sigma^2I)^{-1}(\boldsymbol{z}-\mu) \\ 
    &+\frac{D}{2}\ln(2\pi)+\frac{1}{2}\boldsymbol{z}^{T}\boldsymbol{z} \bigg) d\boldsymbol{z} \\
    &= \frac{1}{2}\int q(\boldsymbol{z}) \bigg(\ln \big( |(\sigma^2I)|^{-1} \big) -(\boldsymbol{z}-\mu)^{T}(\sigma^2I)^{-1}(\boldsymbol{z}-\mu)
    + \boldsymbol{z}^{T}\boldsymbol{z} \bigg) d\boldsymbol{z} \\
    &= \frac{1}{2} \bigg( \ln \big( |(\sigma^2I)|^{-1} \big) +  E[\boldsymbol{z}^{T}\boldsymbol{z}]_{q(\boldsymbol{z})} - E[(\boldsymbol{z}-\mu)^{T}(\sigma^2I)^{-1}(\boldsymbol{z}-\mu)]_{q(\boldsymbol{z})} \bigg) \\
    &= \frac{1}{2} \bigg( \ln \big( |(\sigma^2I)|^{-1} \big) +  \mu^{T}\mu+Tr(\sigma^2I) - (\mu-\mu)^{T}(\sigma^2I)^{-1}(\mu-\mu) - Tr((\sigma^2I)^{-1}(\sigma^2I)) \bigg) \\
    &= \frac{1}{2} \bigg( \ln \big( |(\sigma^2I)|^{-1} \big) +  \mu^{T}\mu + Tr(\sigma^2I) - D \bigg)  \\ 
\end{aligned}
$$

Write a function `KL_loss` that takes two J dimensional vectors `mu` and `logvar` and returns the negative KullbackLeibler divergence. Where `logvar` is $\log(\sigma^2)$. Make sure that your function works for batches of arbitrary size.

````python
def KL_loss(mu, logvar):
    # YOUR CODE HERE
    D = mu.size()[1]
    # due to the input given we do not use 1:1 the derived formula
    return (1/2)*torch.sum(-logvar + mu.pow(2) + logvar.exp()) - D
````

### 2.4 Reparameterization (10 points)
One of the major question that the VAE is answering, is 'how to take derivatives with respect to the parameters of a stochastic variable?', i.e. if we are given $\bz$ that is drawn from a distribution $q(\bz|\bx)$, and we want to take derivatives. This step is necessary to be able to use gradient-based optimization algorithms like SGD.
For some distributions, it is possible to reparameterize samples in a clever way, such that the stochasticity is independent of the parameters. We want our samples to deterministically depend on the parameters of the distribution. For example, in a normally-distributed variable with mean $\mu$ and standard deviation $\sigma$, we can sample from it like this:

\begin{align}
\bz = \mu + \sigma \odot \epsilon,
\end{align}

where $\odot$ is the element-wise multiplication and $\epsilon$ is sampled from $N(0, I)$.


Write a function `reparameterize` that takes two J dimensional vectors `mu` and `logvar`. It should return $\bz = \mu + \sigma \odot \epsilon$.


````python
def reparameterize(self, mu, logvar):
    # returns z
    # YOUR CODE HERE
    noise = torch.randn(mu.size())
    return mu + (torch.sqrt(torch.exp(logvar)) * noise)
````

### 2.6 Forward pass (10)
To complete the data structure you have to define a forward pass through the VAE. A single forward pass consists of the encoding of an MNIST image $\bx$ into latent space $\bz$, the reparameterization of $\bz$ and the decoding of $\bz$ into an image $\bx$.

Write a function `forward` that gets a a vector `x` with 784 elements (flattened MNIST image) and returns a vector `x_hat` with 784 elements (flattened MNIST image), `mu` and `logvar`.

````python
def forward(self, x):
    # YOUR CODE HERE
    x = x.view(-1, 784)
    mu, logvar = self.encode(x)
    z = self.reparameterize(mu, logvar)
    x_hat = self.decode(z)
    
    return x_hat, mu, logvar
````

### 2.7 Training (15)
We will now train the VAE using an optimizer called Adam, https://arxiv.org/abs/1412.6980. The code to train a model in PyTorch is given below.

````python
from torch.autograd import Variable

def train(epoch, train_loader, model, optimizer):
    model.train()
    train_loss = 0
    for batch_idx, (data, _) in enumerate(train_loader):
        data = Variable(data)
        optimizer.zero_grad()
        recon_batch, mu, logvar = model(data)
        loss = loss_function(recon_batch, data.view(-1, 784), mu, logvar)
        loss.backward()
        train_loss += loss.data
        optimizer.step()
        if batch_idx % 100 == 0:
            print('Train Epoch: {} [{}/{} ({:.0f}%)]\tLoss: {:.6f}'.format(
                epoch, batch_idx * len(data), len(train_loader.dataset),
                100. * batch_idx / len(train_loader),
                loss.data / len(data)))

    print('====> Epoch: {} Average loss: {:.4f}'.format(
          epoch, train_loss / len(train_loader.dataset)))
````

### 2.8 Amortized inference (10 points)
What is amortized inference? Where in the code of Part 2 is it used? What is the benefit of using it?



Consider a latent variable model where there is one latent variable corresponding to each observed variable. 
If we want to find the posterior distribution of the latent variable conditioned on the observations, we can make use of variational inference. In the regular case of mean field variational inference, we approximate the posterior of interest using a mixture of simpler distributions (for example Gaussian distributions). This is done by optimizing the hyperparameters of the approximator funtions (i.e. mean and variance in case of the Gaussian function) using an objecive function. However, the problem with mean field variational inference is that for every observation we use one function in the mixture model. This means that the number of hyperparameters that need to be optimized grows linearily with the number of observations considered, which can be problematic if we have a large amount of observations. Further, the approximating mixture model is not generelizable - if we would want to consider a new observation we would need to run the whole optimizition process again while including the new observation.

To approach these drawbacks, amortized inference makes use of a parameterized function which maps from the space of observations to the space of approximate distribution parameters. In case of the VAE, this function is represented by the encoder network $q_{\phi}(z | x)$: It receives an observation $x$ as input and outputs mean and variance of a Gaussian distribution approximating $p(z|x)$. Obviously, this is great in terms of runtime because the number of parameters to tune is not anymore dependent on the number of observations considered. It further allows for generalization: If we want the latent variable corresponding to a new observation $x$ we can just pass $x$ through the encoder network instead of doing the whole optimization again.
